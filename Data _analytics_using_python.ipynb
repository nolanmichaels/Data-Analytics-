## 3 interesting Facts about Campaign Expenditures for NY Candidates

Background on my project: A look at the listing of campaign expenditures for candidates for City office dating back from 2013 election cycle. 
The publisher of the dataset is data.ctyofnewyork.us

```python
import pandas as pd
#Here is a link to my dataset: https://catalog.data.gov/dataset/2013-campaign-expenditures-e8c6d
```

```python
df = pd.read_csv(r"C:\Users\Nolan Michaels\Documents\2013_Campaign_Expenditures.csv")
print(df)
```

```python
import matplotlib.pyplot as plt
```

## Interesting Fact #1: What candidates receive the most donations. 

```python
axis = df['CANDID'].hist(bins=100)
axis.set(xlabel='CANDIDATE NUMBER', ylabel = 'Total Count', title= 'Histogram of Candidate Donation Disbtribution')
#Histogram representing the disribution of the number of donations to each candidate. 
```

```python
df.groupby(['CANDID','AMNT']).count()
#This is a tabular representation of the number of people who donated. 
```

```python
df.plot.scatter(x='CANDID', y = 'AMNT', title = "Scatterplot Between State and Amount Donated")
#The scatterplot shows the amount people are donating to specific candidates.
#Notice that the candidate #1267 has the most donations but not the highest amounts donated.
#unfortunately, the canddiates are listed by numbers so there is not an easy way to identify them.
```

```python
df.groupby(['CANDID']).mean()
#Just looking at the CANDID number and the AMNT column, you can see average amount donated for each candidate.
```

## Interesting Fact #2: What states donate the most on average

```python
ax = df['STATE'].value_counts().plot(kind='barh', 
                                    rot = 0,
                                    figsize = (10,16),
                                    title="Number for each Owner Name")
ax.set_xlabel("STATE")
ax.set_ylabel("Number of Donations from each State")


#https://stackoverflow.com/questions/48939795/how-to-plot-a-count-bar-chart-with-a-pandas-df-grouping-by-one-categorical-colu
```

```python
pd.pivot_table(df,index=["STATE"], values = ['AMNT'], aggfunc ='count')
#here we can see the number of people who donated in each state through a pivot table.
```

```python
pd.pivot_table(df,index=["STATE"], values = ['AMNT'], aggfunc ='mean')
#here we can see the average amount was donated to candidates in each state. 
#It appears that the most was donated on average in IL and then AL (Alabama). 
```

## Interesting Fact #3: Discrepancy in Donors with an original indicator? 

```python
#It should be noted that ORG_IND highlights people who originally indicated they were going to donated. 
```

```python
axis = df['ORG_IND'].value_counts().plot(kind='bar', 
                                    title="Number of People who Originally Indicated They Would Donate")
axis.set_xlabel("Original Indicators")
axis.set_ylabel("Number of People")
#This bar chart visually demonstrates the number of people who marked Y or N to whether 
#the Y column is comprised of people who original indicated they would donate.
```

```python
pd.pivot_table(df,index=['ORG_IND'],values = ['AMNT'], aggfunc = 'mean') 
#Here we can see that people who originally indicated that they were going to donate, donated more on average.
```
